```python
import librosa
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

# Ses dosyasını yükle
# file_path = "sample_angry.wav"
file_path = "sample.wav"


y, sr = librosa.load(file_path, sr=None)

# Mel-spektrogram çıkarımı
mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

# Normalizasyon
mel_spec_db = (mel_spec_db - np.mean(mel_spec_db)) / np.std(mel_spec_db)

# Veri seti (örnekleme için tek dosya yerine çoklu dosyalar kullanılmalı)
X = []
y_labels = []

X.append(mel_spec_db.T)   # Zaman ekseni transpose
y_labels.append("angry")

# Encode etiketler
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y_labels)

# Eğitim / test bölme
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Tensorflow tensörlere dönüştür
X_train = np.array(X_train)
X_test = np.array(X_test)

# CNN + LSTM model
model = models.Sequential([
    layers.Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    layers.MaxPooling1D(pool_size=2),
    layers.LSTM(128, return_sequences=False),
    layers.Dense(64, activation='relu'),
    layers.Dense(len(encoder.classes_), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Eğitim
model.fit(X_train, y_train, epochs=10, batch_size=8, validation_split=0.2)

# Test
y_pred = np.argmax(model.predict(X_test), axis=1)
print(classification_report(y_test, y_pred, target_names=encoder.classes_))
```

```python
import os  # Added missing import
import numpy as np  # Fixed the 'as np' to 'import numpy as np'
import librosa
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

# Veri seti klasörü (RAVDESS)
DATASET_PATH = "ravdess/"

def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=None)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    mel_spec_db = (mel_spec_db - np.mean(mel_spec_db)) / np.std(mel_spec_db)
    return mel_spec_db.T

X, y_labels = [], []

# Add error checking to verify dataset path exists
if not os.path.exists(DATASET_PATH):
    raise FileNotFoundError(f"Dataset path '{DATASET_PATH}' does not exist")

# Add debug print to check files are being found
print(f"Looking for .wav files in {DATASET_PATH}")

for root, _, files in os.walk(DATASET_PATH):
    for file in files:
        if file.endswith(".wav"):
            try:
                file_path = os.path.join(root, file)
                print(f"Processing file: {file_path}")  # Debug print
                
                feature = extract_features(file_path)
                X.append(feature)
                
                # RAVDESS dosya adında duygu kodu var (ör: 03 = happy)
                emotion_code = int(file.split("-")[2])
                emotion_map = {
                    1: "neutral", 2: "calm", 3: "happy", 4: "sad",
                    5: "angry", 6: "fearful", 7: "disgust", 8: "surprised"
                }
                y_labels.append(emotion_map[emotion_code])
            except Exception as e:
                print(f"Error processing file {file}: {e}")

# Check if data was loaded
if len(X) == 0:
    raise ValueError("No data was loaded. Check your dataset path and file structure.")

print(f"Loaded {len(X)} samples")

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y_labels)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Ensure all samples have the same shape before converting to numpy array
# Find the minimum length across all samples
min_length = min(x.shape[0] for x in X_train)
print(f"Minimum sequence length: {min_length}")

# Truncate all sequences to the minimum length
X_train = np.array([x[:min_length] for x in X_train])
X_test = np.array([x[:min_length] for x in X_test])

model = models.Sequential([
    layers.Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    layers.MaxPooling1D(pool_size=2),
    layers.LSTM(128),
    layers.Dense(64, activation='relu'),
    layers.Dense(len(encoder.classes_), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=30, batch_size=16, validation_split=0.2)

y_pred = np.argmax(model.predict(X_test), axis=1)
print(classification_report(y_test, y_pred, target_names=encoder.classes_))
```
